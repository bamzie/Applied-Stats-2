### Problem A.1 (10 points)

Recall the following one-way ANOVA notation and assumptions:

- Assume that there are $J$ groups/levels, with $n_j$ units in each level ($j = 1,...,J$).


- The true mean of the $j^{th}$ group is denoted $\mu_j$.


- $Y_{i,j}$ is the response value for an individual in the $i^{th}$ group at the $j^{th}$ level (for a fixed $j$, $i = 1,...,n_j$)..


- $Y_{i,j} = \mu_j + \varepsilon_{i,j}$, where $\varepsilon_{i,j} \sim N(0,\sigma^2)$.


- $\bar{Y}_{\cdot,j}$ is the sample mean of all units in the $j^{th}$ level.


**Show that, for any set of constants $\mathbf{a} = (a_1,...,a_J)^T$, the quantity $\displaystyle \sum^J_{j=1}a_j\bar{Y}_{\cdot,j}$ is normally distributed with mean $\displaystyle \sum^J_{j=1}a_j\mu_j$ and variance $\displaystyle \sigma^2\sum^J_{j=1}a^2_j\big/n_j$.**

$$
E[\sum^J_{j=1}a_j\bar{Y}_{\cdot,j}] = \sum^j_{j=1}a_jE[\bar{Y}_{.,j}] = \sum^j_{j=1}a_j\mu_j
$$

$$
Var(\sum^J_{j=1}a_j\bar{Y}_{\cdot,j}) = \sum^j_{j=1} a^2_jVar(\bar{Y}_{.,j}) = \sigma^2\sum^j_{j=1}\frac{a^2_j}{n_j}
$$
